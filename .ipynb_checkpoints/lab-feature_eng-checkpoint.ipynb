```python
#Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head()
```

**Check for missing values**

```python
spaceship.info()
```

There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

```python
print(spaceship.isna().sum())
```

```python
df_cleaned = spaceship.dropna()
df_cleaned
```

```python
df_cleaned_Transported= pd.DataFrame(df_cleaned['Transported'])
df_cleaned_Transported
```

- **Cabin** is too granular - transform it in order to obtain {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}

```python
df_cleaned['Cabin'] = df_cleaned['Cabin'].astype(str)

# Extract the first letter of the cabin
df_cleaned['Cabin'] = df_cleaned['Cabin'].str[0]

# Replace 'n' from None with 'NaN' if you prefer to keep missing values as NaN
df_cleaned['Cabin'] = df_cleaned['Cabin'].replace('n', pd.NA)

print("\nTransformed DataFrame:")
print(df_cleaned)
df_cleaned.isna().sum()
```

- Drop PassengerId and Name

```python
spaceship_cleaned= df_cleaned.drop(columns = ["PassengerId","Name"])
spaceship_cleaned
```

```python
spaceship_num = df_cleaned.select_dtypes(include=['number'])

print("Numerical columns in the DataFrame:")
spaceship_num
```

```python
spaceship_obj = spaceship_cleaned.select_dtypes(include=['object'])
spaceship_obj
```

- For non-numerical columns, do dummies.

```python
df_encoded = pd.get_dummies(spaceship_obj, drop_first = True).astype(int)
df_encoded
```

```python
concat_spaceship= pd.concat([spaceship_num, df_encoded, df_cleaned_Transported], axis = 1)
concat_spaceship
```

**Perform Train Test Split**

```python
features = concat_spaceship.drop(columns = ["Transported"])
target = concat_spaceship["Transported"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.20, random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler
normalizer = MinMaxScaler()
normalizer.fit(X_train)
```

```python
X_train_norm = normalizer.transform(X_train)

X_test_norm = normalizer.transform(X_test)
```

```python
X_train_norm = pd.DataFrame(X_train_norm, columns = X_train.columns)
X_test_norm = pd.DataFrame(X_test_norm, columns = X_test.columns)
```

**Model Selection**

In this exercise we will be using **KNN** as our predictive model.

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train_norm, y_train)
```

- Evaluate your model's performance. Comment it

```python
knn.score(X_test_norm, y_test)
```
